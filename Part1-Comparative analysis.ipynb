```python
def check_for_uniqueness(row,brain_part):
    brain_parts = ['Olfactory Bulb','Mid brain','Cerebellum','Hippocampus','Medulla','Hypothalamus','Cerebral cortex']
    brain_parts.remove(brain_part)
    temp = 0
    for i in range(len(brain_parts)):
        temp = temp or row[brain_parts[i]]
    if row[brain_part]==1 and temp ==0:
        return True
    else:
        return False
```

```python
def create_access_name(row):
    s=row["Description"]
    return s[s.find("[")+1:s.find("]")]
  
```

```python
def find_unique(data_given,brain_part):
#     brain_parts = [MB_data_given,OB_data_given,MD_data_given,CB_data_given,HC_data_given,HT_data_given,CC_data_given]
    brain_parts = ["MB","OB","MD","CB","HC","HT","CC"]
    brain_parts.remove(brain_part)
    temp=[]
    for v in brain_parts:
        temp.append(globals()[f'{v}_data_given'])
    for i in range(len(temp)):
    # belong to data_given and doesn't belong to _unique_found
#         temp = int(brain_parts[i]+"_unique_found")
        data_given = data_given[~data_given['Accession Name'].isin(temp[i]['Accession Name'])]
    return data_given
    

```
