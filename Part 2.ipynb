```python
import re
import requests
import urllib

def word_break(text_br):
    words = re.findall(r'\b\S+\b', text_br) 
    return words

def retrieve_prot_inf(query):
    global K
    K=K+1
    print (query,"  ---  ", K, end="\r")
    ac_uniprot = ""
    id_uniprot = ""
    res = urllib.request.urlopen("http://www.uniprot.org/uniprot/" + query + ".txt").read()
    s = str(res, 'utf-8')
#     print(s)
    lines = s.splitlines()
#     print(lines)
#   print(res)
    cellular_comp = [] #C
    molecular_func = [] #F
    biological_proc = [] #P
    cellular_comp_id = [] #C
    molecular_func_id = [] #F
    biological_proc_id = [] #P
    
    for i in range(len(lines)):
        match_word=word_break(lines[i])
        if len(lines[i])>2:
            if match_word[0]=="DR" and match_word[1]=="GO":
#                 print(lines[i])
#                 print(match_word)
#                 print(len(match_word))
#                 print(len(match_word)-5)
#                 print("-------------------------------")
                gene_id = match_word[2]
                
                if match_word[3][0] == "C":
#                     print(match_word[3][2:])
                    sentence = match_word[3][2:]
                    #because we must exclude DR GO GO:342552 and the first and last word
                    for j in range(4,len(match_word)-1):
                        sentence = sentence +" "+ match_word[j]
                    cellular_comp.append(sentence)
                    cellular_comp_id.append(gene_id)
                elif match_word[3][0] == "F":
#                     print(match_word)
#                     print(match_word[3][2:])
                    sentence = match_word[3][2:]
                    #because we must exclude DR GO GO:342552 and the first and last word
                    for j in range(4,len(match_word)-1):
                        sentence = sentence +" "+ match_word[j]
                    molecular_func.append(sentence)
                    molecular_func_id.append(gene_id)
                elif match_word[3][0] == "P":
#                     print(match_word[3][2:])
                    sentence = match_word[3][2:]
                    #because we must exclude DR GO GO:342552 and the first and last word
                    for j in range(4,len(match_word)-1):
                        sentence = sentence +" "+ match_word[j]
                    biological_proc.append(sentence)
                    biological_proc_id.append(gene_id)
                
    return cellular_comp,molecular_func,biological_proc,cellular_comp_id,molecular_func_id,biological_proc_id

```
